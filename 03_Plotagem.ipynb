# Plotagem de gráficos em Python
### Exemplo: solução analítica da viga de Euler-Bernoulli 
Definição das bibliotecas e parâmetros da viga

```python
# Importar bibliotecas
import matplotlib.pyplot as plt
import numpy as np
# Definir valor dos parâmetros da viga
E,I,q,L=50e9,.15*.4**3/12,20e3,4
```

Início da Plotagem

```python
# Definir vetor x contendo pontos no domínio da viga
x = np.linspace(0, 4, 100)
# Plotagem da função v: solução analítica da viga bi-apoiada
plt.plot(x, -q/(24*E*I)*(x**4 - 2*L*x**3 + L**3*x), label='solução analítica')  
plt.xlabel('x')
plt.ylabel('v: deslocamento vertical')
plt.title("Viga bi-apoiada")
plt.legend()
```
